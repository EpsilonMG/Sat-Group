```python
from datasist.structdata import detect_outliers
outlier_indices=detect_outliers(df,0,df.columns)

len(outlier_indices)
```

```python
m=['int.rate','installment','log.annual.inc','fico','days.with.cr.line','revol.bal','revol.util']
for col in m:
    outlier_indecies=detect_outliers(df,0,[col])
    df.loc[outlier_indecies,col]=df[col].median()
```
