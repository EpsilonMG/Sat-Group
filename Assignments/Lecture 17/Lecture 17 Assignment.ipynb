# Houses Prices

```python
df=pd.read_csv('Houses prices.csv')
df
```

```python
def calc_profit(x):
    profit=x['price']*20+x['condition']*5
    maint=2000/(365*24)
    profit=profit-maint
    total_profit=profit-(profit*0.14)
    return total_profit

#df['Profit']=df['casual']*20+df['registered']*5 -(df['casual']*20+df['registered']*5)*0.14 + 1500/(365*24)
df['Profit']=df.apply(calc_profit,axis=1)
df
```

```python
df.info()
```

```python
df['date']=pd.to_datetime(df['date'],format='%Y-%m-%d %H:%M:%S',errors='coerce')
df
```

```python
df.info()
```

```python
df['month']=df['date'].dt.month
df['hour']=df['date'].dt.hour
df['dayname']=df['date'].dt.day_name()
```

```python
df
```

```python
x=df.drop(['date'],axis=1)
x
```

```python
df.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='most_frequent')
df[['street','waterfront','view','city']]=imputer.fit_transform(df[['street','waterfront','view','city']])
df
```

```python
df.isna().sum()
```

```python
from sklearn.impute import KNNImputer
imputer=KNNImputer()
df[['bathrooms','sqft_living' ,'sqft_lot']]=imputer.fit_transform(df[['bathrooms','sqft_living' ,'sqft_lot']])
df
```

```python
df.isna().sum()
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
def pod(x):
    if x in range(0,13):
        return 'Morning'
    elif x in range(13,18):
        return 'AfterNoon'
    elif x in range(18,21):
        return 'Evening'
    else:
        return 'Night'
df['Period']=df['hour'].apply(pod)
df
```

```python
df.drop(['price','sqft_living','sqft_lot','waterfront','view'],axis=1,inplace=True)
df
```

```python
df=pd.get_dummies(df,columns=['bedrooms','floors','condition'],drop_first=True)
df
```

```python
!pip install datasist
```

```python
from datasist.structdata import detect_outliers
indx=detect_outliers(df,0,['sqft_above'])
```

```python
df.drop(indx,inplace=True)
df
```

```python
x=df.drop(['date','street','city','statezip','country'],axis=1)
y=df['bathrooms']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
models={
    'LR':LinearRegression(),
    'KNN':KNeighborsRegressor(n_neighbors=5),
    'SVR':SVR(),
    'DT':DecisionTreeRegressor(),
    'RF':RandomForestRegressor(n_estimators=15),
    'XGBR':XGBRegressor()
}
```

```python
for name,model in models.items():
    print(f'using:{name}')
    model.fit(x_train,y_train)
    print(f'training accuracy: {model.score(x_train,y_train)}')
    print(f'testing accuracy: {model.score(x_test,y_test)}')
    print('_'*30)
```
