```python
#importing libraries
import os
import random
import time

#defining functions:
#1- Accessing files from directory
def show_universe_people(place):
    people = os.listdir(place)
    return people

#2- deleting half the files(randomly selected) 
def thanos_snap(universe):
    half_of_universe = len(universe) // 2
    for i in range(half_of_universe):
        universe = show_universe_people('universe')
        random_victim = random.choice(universe)
        os.remove(f'universe/{random_victim}')
        time.sleep(0.5)
        print(f"\n\nDestroying: {random_victim}")
        
    new_universe = show_universe_people('universe')
    print('\nNew universe population length is: ', len(new_universe))
    print('New universe population is: \n', new_universe)
    print('\nPerfectly balanced, as all things should be')
    
universe = show_universe_people('universe')
print("Current universe population is: \n\n",universe)
thanos_snap(universe)
```
