### bikes.csv

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('Assignment datasets/bikes.csv')
df
```

```python
def ext_profit(x):
    profit=(x['registered']*5)+(x['casual']*20)
    taxes=profit*0.14
    maint=1500/(365*24)
    profit=profit-(taxes+maint)
    return profit

df['Profit']=df.apply(ext_profit,axis=1)
df
```

```python
sns.set(rc={'figure.figsize': [10, 10]}, font_scale=1.3)
```

```python
#Distribution rental_bikes_count & profit

sns.jointplot(x="rented_bikes_count",y="Profit",data=df)
```

```python
#Which season that has more registeration?

sns.barplot(x="registered",y='season',data=df)
```

```python
#Which season that has more profit?

sns.barplot(x="Profit",y='season',data=df)
```

```python
#Which weather condition that has more registeration?

sns.barplot(x="registered",y='weather',data=df)
```

```python
#Which weather condition that has more profit?

sns.barplot(x="Profit",y='weather',data=df)
```

```python
df['datetime']=pd.to_datetime(df['datetime'],format='%Y/%m/%d %H:%M:%S')
df
```

```python
df['rented_bikes_hour']=df['datetime'].dt.hour
df['rented_bikes_month']=df['datetime'].dt.month
df['rented_bikes_day_name']=df['datetime'].dt.day_name()
df
```

```python
#Which hour has more profit?

sns.jointplot(x="rented_bikes_hour",y="Profit",data=df)
```

```python
#Which month has more profit?

sns.jointplot(x="rented_bikes_month",y='Profit',data=df)
```

```python
#Which day has more profit?

sns.barplot(x="Profit",y='rented_bikes_day_name',data=df)
```
