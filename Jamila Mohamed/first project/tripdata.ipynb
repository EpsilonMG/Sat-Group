```python
df['tpep_pickup_datetime']=pd.to_datetime(df['tpep_pickup_datetime'],format='%Y/%m/%d %H:%M:%S')
df['tpep_dropoff_datetime']=pd.to_datetime(df['tpep_dropoff_datetime'],format='%Y/%m/%d %H:%M:%S')
df
```

```python
df['PickupHour']=df['tpep_pickup_datetime'].dt.hour
df['DropoffHour']=df['tpep_dropoff_datetime'].dt.hour
df['Day_Name']=df['tpep_pickup_datetime'].dt.day_name()
df
```

```python
def ext_pickup_period(x):
    if x in range(0,13):
        return 'Morning'
    elif x in range(13,19):
        return 'AfterNoon'
    else:
        return 'Evening'

df['PickupPeriod']=df['PickupHour'].apply(ext_pickup_period)
df
```

```python
def ext_dropoff_period(x):
    if x in range(0,13):
        return 'Morning'
    elif x in range(13,19):
        return 'AfterNoon'
    else:
        return 'Evening'

df['DropoffPeriod']=df['DropoffHour'].apply(ext_dropoff_period)
df
```

```python
df['Time_diff']=df['tpep_dropoff_datetime']-df['tpep_pickup_datetime']
df
```

```python
def time_in_seconds(x):
    return x.seconds

df['Time_in_sec']=df['Time_diff'].apply(time_in_seconds)
df
```

```python
df[df['Time_in_sec']==0]
```

```python
data_error=df[df['Time_in_sec']==0].index
df.drop(data_error,inplace=True)
df
```

```python
df['speed(m/s)']=(df['trip_distance']*1609.34)/(df['Time_in_sec'])
df
```

```python
df['VendorID'].unique()
```

```python
def ext_VendorID(x):
    if x==1:
        return 'Creative Mobile Technologies'
    elif x==2:
        return 'VeriFone Inc.'
    
df['VendorID']=df['VendorID'].apply(ext_VendorID)
df
```

```python
df['RateCodeID'].unique()
```

```python
def ext_RateCodeID(x):
    if x==1:
        return 'Standard rate'
    elif x==2:
        return 'JFK'
    elif x==3:
        return 'Newark'
    elif x==4:
        return 'Nassau or Westchester'
    elif x==5:
        return 'Negotiated fare'
    elif x==6:
        return 'Group ride'
    
df['RateCodeID']=df['RateCodeID'].apply(ext_RateCodeID)
df
```

```python
df['payment_type'].unique()   
```

```python
def ext_payment_type(x):
    if x==1:
        return 'Credit card'
    elif x==2:
        return 'Cash'
    elif x==3:
        return 'No charge'
    elif x==4:
        return 'Dispute'
    elif x==5:
        return 'Unknown'
    elif x==6:
        return 'Voided trip'
    
df['payment_type']=df['payment_type'].apply(ext_payment_type)
df
```
